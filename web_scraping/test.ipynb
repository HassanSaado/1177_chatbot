```python
# url extraction list extraction from sitemat

url = "https://www.1177.se/00sitemap.xml"

response = requests.get(url)

sitemap_text = response.text

soup = BeautifulSoup(sitemap_text, 'lxml-xml')

all_urls = soup.find_all('loc')

#save 10 random urls for later_tests
import random
ten_urls = []

for _ in range(10):
  i = random.randint(0, 2306)
  ten_urls.append(all_urls[i].get_text())

print(ten_urls)
```

```python
relevant_tags = ['h1', 'h2', 'h3', 'li', 'p']

url = ten_urls[0]

response = requests.get(url)

response_text = response.text

from bs4 import BeautifulSoup

# Assuming 'html_content' contains your HTML content
soup = BeautifulSoup(response_text, 'html.parser')

# List of relevant tags
relevant_tags = ['h1', 'h2', 'h3', 'li', 'p']

# Function to check if a tag is one of the relevant tags
def is_relevant_tag(tag):
    return tag.name in relevant_tags

# Find all tags that match the relevant tags
found_tags = soup.find_all(is_relevant_tag)

# Extract and print text from each of the found tags
for tag in found_tags:
    print(tag.get_text(strip=True))

```

```python
import os
import requests
from bs4 import BeautifulSoup

# List of URLs


# Directory to save the text files
subfolder = "scraped_texts"
os.makedirs(subfolder, exist_ok=True)

# Function to scrape a URL
def scrape_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    relevant_tags = ['h1', 'h2', 'h3', 'li', 'p']

    # Extract text from relevant tags
    texts = [tag.get_text(strip=True) for tag in soup.find_all(relevant_tags)]
    return "\n".join(texts)

# Iterate over the URLs, scrape, and save the content
for i, url in enumerate(ten_urls, start=1):
    try:
        text = scrape_url(url)
        # Saving the text in a file
        file_path = os.path.join(subfolder, f"text_{i}.txt")
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(text)
        print(f"Saved text from {url} to {file_path}")
    except Exception as e:
        print(f"Error scraping {url}: {e}")
```
